```python
import numpy as np
import pandas as pd
from tensorflow.keras import layers, models
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, f1_score
import warnings
warnings.filterwarnings("ignore")
```

```python
PATH = "MetroPT1.csv"

df = pd.read_csv(PATH, parse_dates=["timestamp"])
df = df.sort_values("timestamp").reset_index(drop=True)

print(df.shape)
df.head()
```

```python
horizon = 1800  # 30 دقيقة

fault_now = (
    (df["TP3"] < 8.5) |
    (df["Oil_temperature"] > 75) |
    (df["Motor_current"] > 5)
).astype(int)

df["fault_future"] = fault_now.shift(-horizon)

df = df.dropna()
df["fault_future"] = df["fault_future"].astype(int)

print("Fault rate:", df["fault_future"].mean())
```

```python
drop_cols = ["timestamp", "fault_future"]

X = df.drop(columns=drop_cols)
y = df["fault_future"]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
def create_sequences(X, y, seq_len=60):
    Xs, ys = [], []

    for i in range(len(X) - seq_len):
        Xs.append(X[i:i+seq_len])
        ys.append(y.iloc[i+seq_len])

    return np.array(Xs), np.array(ys)

X_seq, y_seq = create_sequences(X_scaled, y, seq_len=60)

print(X_seq.shape)
```

```python
split = int(len(X_seq)*0.8)

X_train = X_seq[:split]
X_test  = X_seq[split:]

y_train = y_seq[:split]
y_test  = y_seq[split:]

print(X_train.shape, X_test.shape)
```

```python
model = models.Sequential()

model.add(layers.LSTM(64, return_sequences=True, input_shape=(X_seq.shape[1], X_seq.shape[2])))
model.add(layers.Dropout(0.2))

model.add(layers.LSTM(32))
model.add(layers.Dropout(0.2))

model.add(layers.Dense(1, activation="sigmoid"))

model.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
history = model.fit(
    X_train, y_train,
    epochs=10,
    batch_size=64,
    validation_split=0.1,
    verbose=1
)
```

```python
proba = model.predict(X_test).flatten()

best_f1 = 0
best_t = 0.5

for t in np.linspace(0.1, 0.5, 10):
    pred = (proba > t).astype(int)
    f1 = f1_score(y_test, pred)

    if f1 > best_f1:
        best_f1 = f1
        best_t = t

print("Best threshold:", best_t)

y_pred = (proba > best_t).astype(int)

print(classification_report(y_test, y_pred))
print("F1:", f1_score(y_test, y_pred))
```
