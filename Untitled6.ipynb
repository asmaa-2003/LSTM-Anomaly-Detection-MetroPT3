```python
import torch
import torch.nn as nn
import torch.optim as optim
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
class PositionalEncoding(nn.Module):
    def __init__(self, d_model, max_len=100):
        super().__init__()
        pe = torch.zeros(max_len, d_model)

        for pos in range(max_len):
            for i in range(0, d_model, 2):
                pe[pos, i] = math.sin(pos / (10000 ** ((2 * i)/d_model)))
                pe[pos, i+1] = math.cos(pos / (10000 ** ((2 * i)/d_model)))

        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)

    def forward(self, x):
        return x + self.pe[:, :x.size(1)]
```

```python
class SelfAttention(nn.Module):
    def __init__(self, d_model):
        super().__init__()
        self.q = nn.Linear(d_model, d_model)
        self.k = nn.Linear(d_model, d_model)
        self.v = nn.Linear(d_model, d_model)

    def forward(self, x):
        Q = self.q(x)
        K = self.k(x)
        V = self.v(x)

        scores = torch.matmul(Q, K.transpose(-2, -1)) / math.sqrt(x.size(-1))
        weights = torch.softmax(scores, dim=-1)

        output = torch.matmul(weights, V)
        return output, weights
```

```python
class TransformerBlock(nn.Module):
    def __init__(self, d_model):
        super().__init__()
        self.attention = SelfAttention(d_model)
        self.norm1 = nn.LayerNorm(d_model)
        self.ff = nn.Sequential(
            nn.Linear(d_model, d_model*4),
            nn.ReLU(),
            nn.Linear(d_model*4, d_model)
        )
        self.norm2 = nn.LayerNorm(d_model)

    def forward(self, x):
        attn_output, weights = self.attention(x)
        x = self.norm1(x + attn_output)

        ff_output = self.ff(x)
        x = self.norm2(x + ff_output)

        return x, weights
```

```python
class TransformerModel(nn.Module):
    def __init__(self, vocab_size, d_model=64):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, d_model)
        self.pos = PositionalEncoding(d_model)
        self.block = TransformerBlock(d_model)
        self.fc = nn.Linear(d_model, vocab_size)

    def forward(self, x):
        x = self.embedding(x)
        x = self.pos(x)
        x, weights = self.block(x)
        x = self.fc(x)
        return x, weights
```

```python
sentences = [
    "i like machine learning",
    "i like deep learning",  "i love artificial intelligence",
    "learning is a continuous process",
    "data is very important",
    "models learn from data",
    "deep learning uses neural networks",
    "machine learning uses data",
    "python is easy to learn",
    "ai is transforming the world",
    "technology is evolving fast",
    "intelligent systems learn from experience",

    "natural language processing is fascinating",
    "text data needs preprocessing",
    "models can understand language",
    "attention mechanism improves performance",
    "transformers are powerful models",
    "bert is a transformer model",
    "gpt generates text",
    "nlp is used in chatbots",
    "language models predict words",
    "deep models require data",

    "learning algorithms improve with data",
    "data analysis is important",
    "big data is everywhere",
    "ai helps solve problems",
    "automation saves time",
    "prediction models are useful",
    "classification is a common task",
    "training requires optimization",
    "loss function measures error",
    "gradient descent updates weights"

]

words = list(set(" ".join(sentences).split()))
word2idx = {w:i for i,w in enumerate(words)}
idx2word = {i:w for w,i in word2idx.items()}

def encode(sentence):
    return [word2idx[w] for w in sentence.split()]

data = [encode(s) for s in sentences]
```

```python
model = TransformerModel(len(words))
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

for epoch in range(200):
    total_loss = 0
    for seq in data:
        x = torch.tensor(seq[:-1]).unsqueeze(0)
        y = torch.tensor(seq[1:]).unsqueeze(0)

        output, _ = model(x)

        loss = criterion(output.view(-1, len(words)), y.view(-1))

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss:.4f}")
```

```python
def plot_attention(weights, sentence):
    tokens = sentence.split()

    sns.heatmap(weights[0].detach().numpy(),
                xticklabels=tokens,
                yticklabels=tokens,
                cmap="viridis")

    plt.title("Attention Map")
    plt.show()

# Test
x = torch.tensor(encode(sentences[0])[:-1]).unsqueeze(0)
_, weights = model(x)

plot_attention(weights, sentences[0])
```

```python
def generate(model, start_word, max_len=5):
    model.eval()
    words_gen = [start_word]

    for _ in range(max_len):
        x = torch.tensor([word2idx[w] for w in words_gen]).unsqueeze(0)
        output, _ = model(x)

        next_word = torch.argmax(output[0, -1]).item()
        words_gen.append(idx2word[next_word])

    return " ".join(words_gen)

print(generate(model, "i"))
```

```python
def predict_next_word(model, input_text):
    model.eval()

    # Tokenisation
    tokens = input_text.split()
    indices = []

    # Vérifier si les mots existent dans le vocabulaire
    for w in tokens:
        if w in word2idx:
            indices.append(word2idx[w])
        else:
            print(f"❌ Le mot '{w}' n'existe pas dans le vocabulaire")
            return

    # Transformer en tenseur
    x = torch.tensor(indices).unsqueeze(0)

    # Prédiction
    output, _ = model(x)

    # Récupérer le mot suivant (dernier token)
    next_idx = torch.argmax(output[0, -1]).item()
    next_word = idx2word[next_idx]

    print(f"👉 Mot prédit : {next_word}")
```

```python
while True:
    text = input("Entrez une phrase (ou 'exit' pour quitter) : ")

    if text.lower() == "exit":
        break

    predict_next_word(model, text)
```
