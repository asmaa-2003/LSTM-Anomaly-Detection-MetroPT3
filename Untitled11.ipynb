```python
# =========================================================
# 🚆 1. DOWNLOAD DATA
# =========================================================
import os
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"

print("⬇️ Downloading dataset...")
!wget -q https://zenodo.org/records/7766691/files/MetroPT2.csv?download=1 -O MetroPT2.csv
print("✅ Download completed")
```

```python
# =========================================================
# 🚆 2. IMPORT LIBRARIES
# =========================================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score, precision_recall_curve

from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, RepeatVector, TimeDistributed, Dense, Dropout

print("✅ Libraries imported successfully")

# SETTINGS
DATA_PATH = "MetroPT2.csv"
CHUNK_SIZE = 50000
TIMESTEPS = 30   # 🔥 مهم
BATCH_SIZE = 64

print(f"✅ TIMESTEPS={TIMESTEPS}, BATCH_SIZE={BATCH_SIZE}")
```

```python
# =========================================================
# 🚆 3. PREPARE FIRST CHUNK + SCALER
# =========================================================
first_chunk = next(pd.read_csv(DATA_PATH, chunksize=CHUNK_SIZE))

# حذف عمود index إذا موجود
if first_chunk.columns[0].lower() in ["unnamed: 0", "index"]:
    first_chunk = first_chunk.drop(first_chunk.columns[0], axis=1)

# اختيار features
feature_cols = first_chunk.select_dtypes(include=[np.number]).columns.tolist()
if "is_anomaly" in feature_cols:
    feature_cols.remove("is_anomaly")

n_features = len(feature_cols)

# 🔥 FIT ONCE ONLY
scaler = StandardScaler()
scaler.fit(first_chunk[feature_cols].values.astype(np.float32))

print("📊 Number of features:", n_features)
print("✅ Scaler fitted once")
```

```python
# =========================================================
# 🚆 4. CREATE SEQUENCES
# =========================================================
def create_sequences(data, timesteps=TIMESTEPS):
    Xs = []
    for i in range(len(data) - timesteps):
        Xs.append(data[i:i+timesteps])
    return np.array(Xs)

print("✅ Sequence function ready")
```

```python
# =========================================================
# 🚆 5. BUILD MODEL (IMPROVED)
# =========================================================
inputs = Input(shape=(TIMESTEPS, n_features))

x = LSTM(128, activation="tanh", return_sequences=True)(inputs)
x = Dropout(0.2)(x)

x = LSTM(64, activation="tanh", return_sequences=False)(x)

x = RepeatVector(TIMESTEPS)(x)

x = LSTM(64, activation="tanh", return_sequences=True)(x)
x = Dropout(0.2)(x)

outputs = TimeDistributed(Dense(n_features))(x)

model = Model(inputs, outputs)
model.compile(optimizer="adam", loss="mae")  # 🔥 أفضل من mse

model.summary()
```

```python
print("\n🚀 Training...")

for i, chunk in enumerate(pd.read_csv(DATA_PATH, chunksize=CHUNK_SIZE)):

    # ⛔️ نوقف بعد عدد معين من chunks
    if i > 20:
        print("🛑 Stopping early (FAST MODE)")
        break

    if chunk.columns[0].lower() in ["unnamed: 0", "index"]:
        chunk = chunk.drop(chunk.columns[0], axis=1)

    chunk["timestamp"] = pd.to_datetime(chunk["timestamp"])
    chunk = chunk.sort_values("timestamp")

    # labeling anomalies
    chunk["is_anomaly"] = 0
    failure_windows = [
        ("2022-06-01", "2022-06-02"),
        ("2022-07-10", "2022-07-11")
    ]
    for start, end in failure_windows:
        mask = (chunk["timestamp"] >= start) & (chunk["timestamp"] <= end)
        chunk.loc[mask, "is_anomaly"] = 1

    feature_cols = chunk.select_dtypes(include=[np.number]).columns.tolist()
    if "is_anomaly" in feature_cols:
        feature_cols.remove("is_anomaly")

    X_chunk = chunk[feature_cols].values.astype(np.float32)

    train_mask = (chunk["timestamp"] < "2022-06-01") & (chunk["is_anomaly"] == 0)
    X_train = X_chunk[train_mask]

    if len(X_train) < TIMESTEPS:
        continue

    X_scaled = scaler.transform(X_train)
    X_seq = create_sequences(X_scaled, TIMESTEPS)

    if len(X_seq) == 0:
        continue

    model.fit(
        X_seq, X_seq,
        epochs=1,
        batch_size=BATCH_SIZE,
        verbose=0
    )

    print(f"✅ Chunk {i+1} trained")

print("\n🎯 Training completed")
```

```python
# =========================================================
# 🚆 7. EVALUATION (SMART THRESHOLD)
# =========================================================

chunks = []
reader = pd.read_csv(DATA_PATH, chunksize=CHUNK_SIZE)

for chunk in reader:
    chunk["timestamp"] = pd.to_datetime(chunk["timestamp"])
    if chunk["timestamp"].max() >= pd.Timestamp("2022-06-01"):
        chunks.append(chunk)
    if len(chunks) >= 3:
        break

sample = pd.concat(chunks)

if sample.columns[0].lower() in ["unnamed: 0", "index"]:
    sample = sample.drop(sample.columns[0], axis=1)

sample = sample.sort_values("timestamp")

# labeling
sample["is_anomaly"] = 0
failure_windows = [
    ("2022-06-01", "2022-06-02"),
    ("2022-07-10", "2022-07-11")
]
for start, end in failure_windows:
    mask = (sample["timestamp"] >= start) & (sample["timestamp"] <= end)
    sample.loc[mask, "is_anomaly"] = 1

y_true = sample["is_anomaly"].values[TIMESTEPS:]

feature_cols = sample.select_dtypes(include=[np.number]).columns.tolist()
if "is_anomaly" in feature_cols:
    feature_cols.remove("is_anomaly")

X_sample = scaler.transform(sample[feature_cols].values.astype(np.float32))
X_seq = create_sequences(X_sample, TIMESTEPS)

pred = model.predict(X_seq, verbose=0)

# 🔥 MAE error
error = np.mean(np.abs(X_seq - pred), axis=(1,2))

# 🔥 BEST THRESHOLD
precision_vals, recall_vals, thresholds = precision_recall_curve(y_true, error)

f1_scores = 2 * (precision_vals * recall_vals) / (precision_vals + recall_vals + 1e-8)
best_threshold = thresholds[np.argmax(f1_scores)]

print("🔥 Best threshold:", best_threshold)

y_pred = (error > best_threshold).astype(int)

precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)
f1 = f1_score(y_true, y_pred)

print("\n📊 Results:")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")
```

```python
# =========================================================
# 🚆 8. VISUALIZATION
# =========================================================
plt.figure(figsize=(14,5))
plt.plot(error, label="Error")
plt.axhline(best_threshold, color='r', linestyle='--', label="Threshold")
plt.legend()
plt.title("Anomaly Detection")
plt.grid()
plt.show()
```
