```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import f1_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("MetroPT3.csv", index_col=0)  # first col is unnamed index
df.head()
```

```python
# Drop timestamp; MetroPT3 has no GPS or Flowmeter columns
df = df.drop(columns=['timestamp'])
df = df.fillna(df.median(numeric_only=True))
df.shape
```

```python
# Define anomaly target: top 1% of Motor_current values
threshold = df['Motor_current'].quantile(0.99)

df['Target'] = (df['Motor_current'] > threshold).astype(int)

df['Target'].value_counts()
```

```python
X = df.drop(['Target', 'Motor_current'], axis=1)
y = df['Target']

# MetroPT3 feature columns:
# TP2, TP3, H1, DV_pressure, Reservoirs, Oil_temperature,
# COMP, DV_eletric, Towers, MPG, LPS, Pressure_switch,
# Oil_level, Caudal_impulses
print(X.columns.tolist())
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.33,
    random_state=42,
    stratify=y
)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
model = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['sqrt', 'log2'],
    'max_depth':    [10, 15, 20, None],
    'criterion':    ['gini', 'entropy']
}

CV_model = GridSearchCV(model, param_grid, cv=5, scoring='f1')
CV_model.fit(X_train, y_train)

CV_model.best_params_
```

```python
best = CV_model.best_params_

mdl = RandomForestClassifier(
    n_estimators=best['n_estimators'],
    max_features=best['max_features'],
    max_depth=best['max_depth'],
    criterion=best['criterion'],
    random_state=42
)

mdl.fit(X_train, y_train)
```

```python
y_pred = mdl.predict(X_test)

print("F1 Score:", f1_score(y_test, y_pred))
print(classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm).plot()
plt.title("Random Forest - MetroPT3")
plt.show()
```

```python
importances = mdl.feature_importances_

feat_df = pd.DataFrame({
    "Feature":    X.columns,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

print(feat_df)

plt.barh(feat_df["Feature"], feat_df["Importance"])
plt.title("Feature Importance - MetroPT3")
plt.show()
```
