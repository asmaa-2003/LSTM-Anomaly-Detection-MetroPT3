```python
# =========================================================
# 🚆 1. DOWNLOAD DATA DIRECTLY IN COLAB
# =========================================================
import os
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"

DATA_PATH = "MetroPT2.csv"

print("⬇️ Downloading dataset...")
!wget -q https://zenodo.org/records/7766691/files/MetroPT2.csv?download=1 -O {DATA_PATH}
print("✅ Download completed")
```

```python
# =========================
# 🔥 2. ENV SETUP
# =========================

import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score, precision_recall_curve
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, RepeatVector, TimeDistributed, Dense, Attention

print("✅ Libraries loaded")
```

```python
# =========================
# 🔥 3. LOAD & PREP DATA
# =========================

CHUNK_SIZE = 10000
TIMESTEPS = 30  # طول التسلسل

chunks = []
reader = pd.read_csv(DATA_PATH, chunksize=CHUNK_SIZE)

for chunk in reader:
    chunk["timestamp"] = pd.to_datetime(chunk["timestamp"])
    if chunk["timestamp"].max() >= pd.Timestamp("2022-06-01"):
        chunks.append(chunk)
    if len(chunks) >= 5:  # استخدمي 5 chunks للتجربة
        break

sample = pd.concat(chunks)

# إزالة عمود index إذا موجود
if sample.columns[0].lower() in ["unnamed: 0", "index"]:
    sample = sample.drop(sample.columns[0], axis=1)

sample = sample.sort_values("timestamp")

# Labeling
sample["is_anomaly"] = 0
failure_windows = [
    ("2022-06-01", "2022-06-02"),
    ("2022-07-10", "2022-07-11")
]
for start, end in failure_windows:
    mask = (sample["timestamp"] >= start) & (sample["timestamp"] <= end)
    sample.loc[mask, "is_anomaly"] = 1

y_true = sample["is_anomaly"].values[TIMESTEPS:]
print("✅ Data loaded and labeled")
```

```python
# =========================
# 🔥 4. FEATURES & SCALER
# =========================

feature_cols = sample.select_dtypes(include=[np.number]).columns.tolist()
if "is_anomaly" in feature_cols:
    feature_cols.remove("is_anomaly")

scaler = StandardScaler()
X_sample = scaler.fit_transform(sample[feature_cols].values.astype(np.float32))

# دالة لإنشاء sequences
def create_sequences(data, timesteps=30):
    sequences = []
    for i in range(len(data) - timesteps):
        sequences.append(data[i:i+timesteps])
    return np.array(sequences, dtype=np.float32)

X_seq = create_sequences(X_sample, TIMESTEPS)
print("✅ Features scaled and sequences created, shape:", X_seq.shape)
```

```python
# =========================
# 🔥 5. LSTM + ATTENTION MODEL
# =========================

inputs = Input(shape=(TIMESTEPS, X_seq.shape[2]))

# Encoder
encoded = LSTM(64, return_sequences=True)(inputs)

# Attention
attention = Attention()([encoded, encoded])

# Encoder آخر
encoded = LSTM(32)(attention)

# Decoder
decoded = RepeatVector(TIMESTEPS)(encoded)
decoded = LSTM(64, return_sequences=True)(decoded)
outputs = TimeDistributed(Dense(X_seq.shape[2]))(decoded)

model = Model(inputs, outputs)
model.compile(optimizer='adam', loss='mae')

model.summary()
```

```python
# =========================
# 🔥 6. TRAINING
# =========================

EPOCHS = 1  # جربي 1 أولاً
BATCH_SIZE = 128

history = model.fit(
    X_seq, X_seq,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    validation_split=0.1,
    verbose=1
)
```

```python
# =========================
# 🔥 7. MAE & BEST THRESHOLD
# =========================

pred = model.predict(X_seq, verbose=0)
error = np.mean(np.abs(X_seq - pred), axis=(1,2))

precision_vals, recall_vals, thresholds = precision_recall_curve(y_true, error)
f1_scores = 2 * (precision_vals * recall_vals) / (precision_vals + recall_vals + 1e-8)

best_threshold = thresholds[np.argmax(f1_scores)]
print("🔥 Best threshold:", best_threshold)
```

```python
# =========================
# 🔥 8. SMART THRESHOLD
# =========================

best_f1 = 0
best_threshold_final = 0

for t in thresholds:
    y_pred_temp = (error > t).astype(int)
    f1_temp = f1_score(y_true, y_pred_temp, zero_division=0)

    if f1_temp > best_f1:
        best_f1 = f1_temp
        best_threshold_final = t

# النتائج النهائية
y_pred = (error > best_threshold_final).astype(int)
precision = precision_score(y_true, y_pred)
recall = recall_score(y_true, y_pred)

print("\n🔥 OPTIMIZED RESULTS:")
print(f"Best Threshold: {best_threshold_final:.6f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {best_f1:.4f}")
```
